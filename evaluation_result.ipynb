```python
import pandas as pd
from docx import Document
from docx.shared import Inches

# Data untuk tabel
models = ['Decision Tree', 'Multilayer Perceptron', 'SVM', 'KNN', 'Logistic Regression', 'Random Forest', 'Naive Bayes']
accuracy_1 = [0.877067, 0.886222, 0.887822, 0.759378, 0.877244, 0.899378, 0.870222]
precision_1 = [0.877497, 0.885283, 0.883860, 0.796041, 0.876841, 0.897640, 0.870246]
recall_1 = [0.866414, 0.877832, 0.883702, 0.716431, 0.867420, 0.893137, 0.859273]
f1_score_1 = [0.870909, 0.881059, 0.883780, 0.722586, 0.871345, 0.895201, 0.863705]

# Membuat dataframe
df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracy_1,
    'Precision': precision_1,
    'Recall': recall_1,
    'F1-Score': f1_score_1,
})

# Membuat dokumen Word
doc = Document()

# Menambahkan judul
doc.add_heading('Model Evaluation Report', level=1)

# Menambahkan tabel
table = doc.add_table(rows=1, cols=len(df.columns))

# Menambahkan header tabel
hdr_cells = table.rows[0].cells
for i, col in enumerate(df.columns):
    hdr_cells[i].text = col

# Menambahkan baris dengan data
for index, row in df.iterrows():
    row_cells = table.add_row().cells
    row_cells[0].text = row['Model']
    row_cells[1].text = str(row['Accuracy'])
    row_cells[2].text = str(row['Precision'])
    row_cells[3].text = str(row['Recall'])
    row_cells[4].text = str(row['F1-Score'])

# Menyimpan dokumen
doc.save('model_evaluation_report.docx')
```
